```python
#!/usr/bin/env python3
"""
extract_challenges.py

usage:
    python extract_challenges.py PBP_RAW_20250208.csv challenges.csv
description:
    reads raw nba pbp csv, filters out coach’s‑challenge events,
    and writes a csv with one row per challenge:
        game_id, team_id, quarter, game_clock,
        successful (y/n),
        first_or_second,
        challenged_play_type
"""

import sys
import pandas as pd
import re
from collections import defaultdict

# ---------- helpers ---------------------------------------------------------

RESULT_RE = re.compile(r"\((?:CALL )?(STANDS|OVERTURNED)\)", re.I)
TEAM_RE   = re.compile(r"\[([A-Za-z\s]+?)\]")  # grabs "[Pistons]" etc.

EVENTTYPE_MAP = {
    1: "made shot",
    2: "missed shot",
    3: "free throw",
    4: "rebound",
    5: "turnover",
    6: "foul",
    7: "violation",
    8: "sub",
    9: "timeout",
    10: "jump ball",
    11: "ejection",
    12: "start/period",
    13: "end/period",
    18: "review"      # nba stats convention
}

def parse_success(text: str) -> str:
    m = RESULT_RE.search(text)
    if not m:
        return ""
    return "y" if m.group(1).upper() == "OVERTURNED" else "n"

def parse_team(text: str, team_id_lookup: dict) -> int | None:
    m = TEAM_RE.search(text)
    if not m:
        return None
    name = m.group(1).strip().lower()
    return team_id_lookup.get(name)

def get_play_type(prev_rows: list[pd.Series]) -> str:
    """
    look back up to two rows; return human string of challenged play
    """
    for row in prev_rows[:2]:
        et = row.EVENTMSGTYPE
        desc = (row.HOMEDESCRIPTION
                or row.VISITORDESCRIPTION
                or row.NEUTRALDESCRIPTION
                or "")
        if et in EVENTTYPE_MAP:
            # special‑case shooting fouls etc.
            if et == 6 and "shooting" in desc.lower():
                return "shooting foul"
            return EVENTTYPE_MAP[et]
    return ""

# ---------- main ------------------------------------------------------------

def main(inp: str, out: str) -> None:
    df = pd.read_csv(inp, low_memory=False)

    # friendly access to description text
    df["TEXT"] = df.HOMEDESCRIPTION.fillna("") + \
                 df.VISITORDESCRIPTION.fillna("") + \
                 df.NEUTRALDESCRIPTION.fillna("")

    challenge_mask = df.TEXT.str.contains("CHALLENGE", case=False, na=False)
    # build quick name->team_id map from first row each team appears
    team_id_lookup: dict[str,int] = {}
    for _, r in df.iterrows():
        for col in ("PLAYER1_TEAM_ID", "PLAYER2_TEAM_ID", "PLAYER3_TEAM_ID"):
            tid = r[col]
            if pd.notna(tid):
                team_id_lookup.setdefault(str(r[col+"_CITY"]).lower(), int(tid))
    # store challenge counts
    seen_counts: defaultdict[tuple[int,int], int] = defaultdict(int)

    rows_out = []
    for idx in df[challenge_mask].index:
        row = df.loc[idx]
        game_id = int(row.GAME_ID)
        quarter = int(row.PERIOD)
        clock   = row.PCTIMESTRING

        text = row.TEXT
        successful = parse_success(text)

        # who is challenging?
        # try team_id from bracket text then fallback to player1_team_id
        team_id = parse_team(text, team_id_lookup)
        if team_id is None:
            team_id = row.PLAYER1_TEAM_ID if pd.notna(row.PLAYER1_TEAM_ID) else None

        # first or second?
        key = (game_id, team_id)
        seen_counts[key] += 1
        first_second = "first" if seen_counts[key] == 1 else "second"

        # look at previous two plays
        prev_rows = df.loc[idx-2:idx-1][::-1].to_dict("records")
        play_type = get_play_type([pd.Series(r) for r in prev_rows])

        rows_out.append({
            "game_id": game_id,
            "team_id": team_id,
            "quarter": quarter,
            "game_clock": clock,
            "successful": successful,
            "first_or_second": first_second,
            "challenged_play_type": play_type
        })

    out_df = pd.DataFrame(rows_out)
    out_df.to_csv(out, index=False)
    print(f"wrote {len(out_df)} rows to {out}")

if __name__ == "__main__":
    if len(sys.argv) != 3:
        sys.exit("usage: python extract_challenges.py INFILE.csv OUTFILE.csv")
    main(sys.argv[1], sys.argv[2])
```
